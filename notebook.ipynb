```python
#Load the dataset using `pd.read_csv()`
import pandas as pd
df = pd.read_csv("Placement_Data_Full_Class.csv")
```

```python
#Display first 5 data
df.head()
```

```python
#Check missing values for every column
print(df.isnull().sum())
```

```python
# Fill missing values in the 'salary' column with 0
# Students who were not placed do not have a salary, so replacing with 0 is appropriate.
df['salary'] = df['salary'].fillna(0)
```

```python
# Remove duplicate records (if any)
df.drop_duplicates(inplace=True)

print("Number of duplicate records:", df.duplicated().sum())
```

```python
# Clean text columns by removing extra spaces and standardizing the text format
text_columns = ['gender', 'ssc_b', 'hsc_b', 'hsc_s',
                'degree_t', 'workex', 'specialisation', 'status']

for column in text_columns:
    df[column] = df[column].str.strip().str.title()
```

```python
# Convert student serial number to integer
df['sl_no'] = df['sl_no'].astype(int)

# Convert percentage columns to float
percentage_columns = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']

for column in percentage_columns:
    df[column] = df[column].astype(float)
```

```python
# Rename columns for better readability and consistency
df.rename(columns={
    'sl_no': 'student_id',
    'ssc_p': 'ssc_percentage',
    'ssc_b': 'ssc_board',
    'hsc_p': 'hsc_percentage',
    'hsc_b': 'hsc_board',
    'hsc_s': 'hsc_stream',
    'degree_p': 'degree_percentage',
    'degree_t': 'degree_type',
    'etest_p': 'employability_test_percentage',
    'mba_p': 'mba_percentage'
}, inplace=True)

print(df.columns)
```

```python
# Create a new column for the average academic score
df['average_academic_score'] = (
    df['ssc_percentage'] +
    df['hsc_percentage'] +
    df['degree_percentage'] +
    df['mba_percentage']
) / 4

# Display the new column
df[['student_id', 'average_academic_score']].head()
```

```python
# Save the cleaned dataset as a CSV file
df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully as 'cleaned_dataset.csv'")
```
